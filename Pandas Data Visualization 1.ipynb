Use **df3** to replicate the following plots.

### SCATTER PLOT

```python
df3.plot.scatter(x='a',y='b',c='red',s=35, figsize=(12,3));
```

** HISTOGRAM**

```python
df3['a'].plot.hist();
```

**Use style sheets to set the style to 'ggplot' and redo the histogram from above. Also figure out how to add more bins to it.***

```python
plt.style.use('ggplot')
```

```python
df3['a'].plot.hist(alpha=0.4,bins=25);
```

** Create a boxplot comparing the a and b columns.**

```python
df3[['a','b']].plot.box();
```

** Create a kde plot of the 'd' column **

```python
df3['d'].plot.kde(lw=4);
```

** Figure out how to increase the linewidth and make the linestyle dashed. (Note: You would usually not dash a kde plot line)**

```python
df3['d'].plot.density(lw=5,ls='--')
```

** Create an area plot of all the columns for just the rows up to 30. (hint: use .ix).**

```python
df3.ix[0:30].plot.area(alpha=0.4);
```

** Display the legend outside of the plot**

```python
f = plt.figure()
df3.ix[0:30].plot.area(alpha=0.4,ax=f.gca())
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
